```python
ncols=6
nrows=3
aspect=1.0
import matplotlib.gridspec as gridspec

# fig, axs = plt.subplots(ncols=ncols, 
#                         nrows=nrows, 
#                         sharey=True, 
#                         sharex=True)
# #fig.subplots_adjust(hspace=0.1, left=0.1, right=0.7)

fig = plt.figure()
gs1 = gridspec.GridSpec(nrows, ncols, width_ratios=[1,1,1,1,1,1])

#plt.setp(gs1, aspect=1.0, adjustable='box-forced')

for i in range(0,(ncols*nrows)):
            

           
    x = phot_modesub.values[:,i]


    y = planck_mw['AME'].values[:]

    x_ = x[(x>0) & (y>0)]
    y_ = y[(x>0) & (y>0)]


    xmin = 5e-5#x_.min()
    xmax = 400 #x_.max()
    ymin = 0.01#y_.min()
    ymax = y_.max()

    ax = fig.add_subplot(gs1[i],sharex=ax, sharey=ax)
    #ax.set_aspect(aspect, adjustable='box')

    hb = ax.hexbin(
           x_,
           y_, 
           mincnt=1,
           gridsize=10,
           bins='log', 
           cmap='inferno_r',
           xscale='log',
           yscale='log')


    ax.axis([xmin, xmax, ymin, ymax])

    ax.text(0.3, 0.8,phot_modesub.columns[i], horizontalalignment='center',
      verticalalignment='center',
      transform=ax.transAxes, 
      fontsize=12)

    ax.grid(True)

    ax.set_frame_on(True)

            
# ax = axs[1,0]
# ax.set_ylabel('AME [$\mu{}K_{CMB}$]', fontsize=15)
# ax = axs[-1,2]
# ax.set_xlabel('IR [MJy/sr]', fontsize=15)

# plt.tight_layout(pad=0, w_pad=0, h_pad=0)

plt.show()

fig.savefig("../Plots/AMEvsDust_allsky_allbands.pdf", bbox_inches='tight')  
```
